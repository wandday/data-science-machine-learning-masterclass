```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#Scatter plot - shows relationship between two continous features
```

```python
df = pd.read_csv('dm_office_sales.csv')
```

```python
df.head()
```

```python
plt.figure(figsize=(8, 5), dpi=100)
sns.scatterplot(x='salary', y='sales', data=df, hue='level of education')
```

```python
#specifying the point size based on salary column values
plt.figure(figsize=(8, 5), dpi=100)
sns.scatterplot(x='salary', y='sales', data=df, size='salary')
```

```python
#Making all the markers(point) larger and same size
plt.figure(figsize=(8, 5), dpi=100)
sns.scatterplot(x='salary', y='sales', data=df, s=200, alpha=0.2) #setting alpha parameter for transparency between 0 an 1
```

```python
#Applying style to categorical data
plt.figure(figsize=(8, 5), dpi=100)
sns.scatterplot(x='salary', y='sales', data=df, s=200, style='level of education', hue='level of education')

#Saving the plot
plt.savefig('my_plot.png')
```

```python
#Distribution plot - display a single continuous feature and help visualize properties such as deviation and average values such as the mean.

#there are three main distribution plot types: 
#RUG plot,
#histogram,
#KDE plot which stands for Kernel Density Estimation.
```

```python
sns.rugplot(x='salary', data=df, height=0.5); #Y-axis is not important here
```

```python
sns.set(style='darkgrid') #dark, darkgrid, ticks
sns.displot(x='salary', data=df, bins=20, color='red',edgecolor='blue' ); #can also use histplot
```

```python
sns.set(style='darkgrid')
sns.displot(x='salary', data=df, bins=20, kde=True, rug=True ); #adding KDE and Rug plot
```

```python
#KDE
```

```python
sns.kdeplot(data=df, x='salary');
```

```python
#Count plot is visualization form of value count
```

```python
plt.figure(figsize=(10,4), dpi=200)
sns.countplot(data=df, x='level of education', hue='division', palette='Set2'); #Check documentation for color palettes
```

```python
#Distribution with categories:
#Boxplot - display distribution of a continious variable
#Violinplot
#Swarmplot
#Boxenplot
```

```python
df = pd.read_csv('StudentsPerformance.csv')
```

```python
df.head()
```

```python
#Boxplot - plot a continuos feature against a categorical one
```

```python
plt.figure(figsize=(10,4), dpi=200)
sns.boxplot(data=df, y='reading score', x='test preparation course');
```

```python
#Violinplot
```

```python
plt.figure(figsize=(10,4), dpi=200)
sns.violinplot(data=df, y='reading score', x='parental level of education');
```

```python
#Comparism plot - 2D version of other plot:
#Two types:
#jointplot (histogram + scatterplot )
#pairplot
```

```python
sns.jointplot(data=df, y='reading score', x='math score', kind='hex');
```

```python
sns.jointplot(data=df, y='reading score', x='math score',  hue='gender');
```

```python
#Gridplot - Create a grid base off the categorical column
```

```python
g = sns.PairGrid(df, hue="gender", palette="viridis",hue_kws={"marker": ["o", "+"]})
g = g.map_upper(sns.scatterplot, linewidths=1, edgecolor="w", s=40)
g = g.map_diag(sns.histplot)
g = g.map_lower(sns.kdeplot)
g = g.add_legend();
```

```python
#Matrix plot - visualize all the numeric value in a dataframe
```

```python
#Two types of matrix plot:
#Heatmap
#Clustermap
```

```python
df = pd.read_csv('country_table.csv')
```

```python
df
```

```python
df = df.set_index('Countries')
```

```python
df
```

```python
plt.figure(figsize=(10,4), dpi=200)

#drop life expectancy because it has a different unit. Other columns are rate/percentage
sns.heatmap(df.drop('Life expectancy', axis=1), linewidth=0.5, annot=True);
```

```python
plt.figure(figsize=(10,4), dpi=200)

#Using a diferent colormap
sns.heatmap(df.drop('Life expectancy', axis=1), linewidth=0.5, annot=True, cmap='viridis');
```

```python
plt.figure(figsize=(10,4), dpi=200)

sns.clustermap(df.drop('Life expectancy', axis=1), linewidth=0.5, annot=True, cmap='viridis', col_cluster=False);
```
