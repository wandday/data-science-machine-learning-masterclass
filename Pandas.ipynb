```python
def quality(total_bill, tip):
    if tip/total_bill > 0.25:
        return 'Generous'
    else:
        return 'Other'
```

```python
#Using Lambda expression for the above
df['quality'] = df[['total_bill', 'tip']].apply(lambda df: quality(df['total_bill'], df['tip']), axis=1)
```

```python
#Using Vectorize for the above - Much simpler syntax
df['quality'] = np.vectorize(quality)(df['total_bill'], df['tip'])
```

```python
#Combining dataframes - Inner merge
#On - the column present in the two tables upon with the merge will be perform
#How - when how = Inner, this returns the set of records that is present in the two tables
```

```python
registrations = pd.DataFrame({'reg_id':[1,2,3,4],'name':['Andrew','Bobo','Claire','David']})
logins = pd.DataFrame({'log_id':[1,2,3,4],'name':['Xavier','Andrew','Yolanda','Bobo']})
```

```python
registrations
```

```python
logins
```

```python
#help(pd.merge)
```

```python
#Inner join return only the values present in both tables 
pd.merge(registrations, logins, how='inner', on='name')
```

```python
#Left and Right Merge
```

```python
#The order of the table been provided matters here, the first table is the left while the second table is the right
```

```python
pd.merge(registrations, logins, how='left', on='name') #Returns all values on the left table
```

```python
pd.merge(registrations, logins, how='right', on='name') #Returns all values on the right table
```

```python
#Outer merge - Returns all the values in the two tables
pd.merge(registrations, logins, how='outer', on='name')
```

```python
#Joining on index
```

```python
#First, set name column as index
registrations = registrations.set_index('name')
registrations
```

```python
logins
```

```python
#Tells pandas to perform the join using the left table index(name) and the name column on the right table
pd.merge(registrations, logins, left_index= True, right_on='name',  how='inner') 
```

```python
#Joining when the column names in the two tables are different
```

```python
registrations = registrations.reset_index()
registrations
```

```python
registrations.columns = ['reg_name', 'reg_id'] #Changing the column names
registrations
```

```python
logins
```

```python
pd.merge(registrations, logins, left_on='reg_name', right_on='name',  how='inner') #Specifying the column names
```

```python
#Two prevent repetition of same column, we can drop one of the column (reg_name or name)
results = pd.merge(registrations, logins, left_on='reg_name', right_on='name',  how='inner')
results.drop('reg_name', axis= 1)
```
