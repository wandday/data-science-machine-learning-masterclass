```python
#Figures and Axes
```

```python
a = np.linspace(0,10,11)
b = a ** 4
```

```python
a
```

```python
b
```

```python
x = np.arange(0,10)
y = 2 * x
```

```python
x
```

```python
y
```

```python
# Create Figure (empty canvas)
fig = plt.figure()

# Add set of axes to figure
axes = fig.add_axes([0, 0, 1, 1]) # left, bottom, width, height (range 0 to 1)

# Plot on that set of axes
axes.plot(a, b)

plt.show()
```

```python
# Creates blank canvas
fig = plt.figure()

axes1 = fig.add_axes([0, 0, 1, 1]) # Full figure
axes2 = fig.add_axes([0.2, 0.5, 0.25, 0.25]) # Smaller figure
axes3 = fig.add_axes([1, 1, 0.25, 0.25]) # Starts at top right corner!

# Larger Figure Axes 1
axes1.plot(a, b)

# Use set_ to add to the axes figure
axes1.set_xlabel('X Label')
axes1.set_ylabel('Y Label')
axes1.set_title('Big Figure')

# Insert Figure Axes 2
axes2.plot(a,b)
axes2.set_xlim(8,10)
axes2.set_ylim(4000,10000)
axes2.set_xlabel('X')
axes2.set_ylabel('Y')
axes2.set_title('Zoomed In');

# Insert Figure Axes 3
axes3.plot(a,b)
```

```python
fig = plt.figure()

axes1 = fig.add_axes([0, 0, 1, 1])

axes1.plot(a,b)
axes1.set_xlabel('X')

# bbox_inches ='tight' automatically makes sure the bounding box is correct
fig.savefig('figure.png',bbox_inches='tight')
```

```python
#Subplot Functionality
```

```python
a = np.linspace(0,10,11)
b = a ** 4
```

```python
a
```

```python
b
```

```python
x = np.arange(0,10)
y = 2 * x
```

```python
x
```

```python
y
```

```python
fig,axes = plt.subplots(nrows=1,ncols=2) #This returns a numpy array

axes[0].plot(a,b)
axes[1].plot(x,y)
```

```python
type(axes)
```

```python
# NOTE! This returns 2 dimensional array
fig,axes = plt.subplots(nrows=2,ncols=2)

axes[0][0].plot(a,b)
axes[1][1].plot(x,y)
```

```python
#Matplotlib Styling -Legend
```

```python
x = np.linspace(0,11,10)
```

```python
fig = plt.figure()

ax = fig.add_axes([0,0,1,1])

ax.plot(x, x, label='x vs x' )
ax.plot(x, x**2, label='x vs x^2')
ax.legend(loc=0) #matplotlib will decide the best location to put the legend
```

```python
#Matplotlib Styling -Colours and Styles
```

```python
x = np.linspace(0,11,10)
```

```python
fig = plt.figure()

ax = fig.add_axes([0,0,1,1])

ax.plot(x, x, color='#e6c017',lw=3, ls='--', marker='o') #lw=line width, ls=linestyle, other marker= +, v, s etc
ax.plot(x, x+1, color='purple', linewidth=2, linestyle='-.')
```
